## k Nearst Neighbors Imputer

```python
 # load the data
df = sns.load_dataset('titanic')

df.isnull().sum().sort_values(ascending= False)
```

```python
# impute missing value with knn
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=2)

df['age'] = imputer.fit_transform(df[['age']])

df.isnull().sum().sort_values(ascending= False)
```

# 4 Random forest for imputing missing value


```python
# load the data
df = sns.load_dataset('titanic')

df.isnull().sum().sort_values(ascending= False)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix,r2_score
from sklearn.impute import SimpleImputer
# load the data
df = sns.load_dataset('titanic')

# check the missing value
df.isnull().sum().sort_values(ascending= False)
```

```python
df.drop(['deck'], axis=1, inplace=True)
df.isnull().sum().sort_values(ascending= False)
```

```python
df.head()
```

```python
# encode the data
from sklearn.preprocessing import LabelEncoder
columns = ['sex', 'embarked','who','class','embark_town','alive',]
Label_encoder ={}

for col in columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    Label_encoder[col] = le

df.head()

```

```python
# split the data with missing value and without missing value
df_missing = df[df['age'].isnull()]

df_not_missing = df[df['age'].notnull()]
```

```python
print("The original data shape: ", df.shape)
print("The missing data shape: ", df_missing.shape)
print("The not missing data shape: ", df_not_missing.shape)
```

```python
df_missing.head()
```

```python
df_not_missing.head()
```

```python
print(df.columns)
```

```python
x = df_not_missing.drop(['age'], axis=1)
y = df_not_missing['age']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# random forest
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
rf_model = RandomForestRegressor(n_estimators=100, random_state=0)
rf_model.fit(x_train, y_train)

# evalute the model
y_pred = rf_model.predict(x_test)
print("RMSE FOR RANDOM FOREST:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R2 SCORE FOR RANDOM FOREST:", r2_score(y_test, y_pred))
print("MAE FOR RANDOM FOREST:", mean_absolute_error(y_test, y_pred))
print("MAPE FOR RANDOM FOREST:", mean_absolute_percentage_error(y_test, y_pred))

```

```python
df_missing.isnull().sum().sort_values(ascending = False)
```

```python
# prdict the missing value
y_pred = rf_model.predict(df_missing.drop(['age'], axis=1))
y_pred
```

```python
#replace the missing value
df_missing['age'] = y_pred

df_missing.isnull().sum().sort_values(ascending = False)
```

```python
# combine the data
df = pd.concat([df_missing, df_not_missing], axis=0)
df.shape
df.head()
```

```python
for col in columns:
    
    le = Label_encoder[col]
    
    df[col] = le.inverse_transform(df[col])
    
df.head()
```
