```python
import pandas as pd
```

```python
data = {
    'date': ['2022-11-10','2021/11/10','5/11/2021','2021-11-10'],
    'country':['USA','U.S.A','UNITED STATES','UNITED STATES OF AMERICA'],
    'name' : ['John Doe','Jane Doe','John Smith','Jane Smith'],
    'sales_20222':[100,200,None,400],
    'sales_2021':[100,200,None,400],
    
}
```

```python
df = pd.DataFrame (data)
df.head()
```

```python
# standardize the date column
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
df.head()
```

```python
# impute the date column
df['date'] = df['date'].fillna(df['date'].mode()[0])
```

```python
df.head()
```

```python
# hormonize the country column
country_map = {
    'USA':'United States',
    'U.S.A':'United States',
    'UNITED STATES':'United States',
    'UNITED STATES OF AMERICA':'United States'
}
df['country'] = df['country'].replace(country_map)
df.head()
```

```python
# correct the name column
df['name'] = df['name'].replace({'John Smith':'John Doe','Jane Smith':'Jane Doe'})
df.head()
```

```python
# remove the duplicate names
df = df.drop_duplicates(subset='name')
df.head()
```
