```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
#load the dataset
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.show()
```

```python
 # split the data
x = df.drop('species', axis = 1)
y = df['species']

# encode the target variable
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
# train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
# train the decisionn model
model = DecisionTreeClassifier()
model.fit(x_train, y_train)

# make predictions
y_pred = model.predict(x_test)

# evaluate the model
# print(classification_report(y_test, y_pred))
# print(confusion_matrix(y_test, y_pred))

print('Accuaracy score:', accuracy_score(y_test, y_pred))
print("precision_score:", precision_score(y_test, y_pred, average = 'micro'))
print("recall_score:",recall_score(y_test, y_pred, average = 'micro'))
print("f1_score:" , f1_score(y_test, y_pred, average = 'micro'))
```

```python
%%time
# train the random forest calssifier
rf = DecisionTreeClassifier()
rf.fit(x_train, y_train)

# make predictions
y_pred = rf.predict(x_test)

# evaluate the model
print('Accuaracy score:', accuracy_score(y_test, y_pred))
print("precision_score:", precision_score(y_test, y_pred, average = 'micro'))
print("recall_score:",recall_score(y_test, y_pred, average = 'micro'))
print("f1_score:" , f1_score(y_test, y_pred, average = 'micro'))
```

```python
%%time
# train the xgboost model
xgb_model = XGBClassifier()
xgb_model.fit(x_train, y_train)

# make predictions
y_pred = xgb_model.predict(x_test)

# evaluate the model
print('Accuaracy score:', accuracy_score(y_test, y_pred))
print("precision_score:", precision_score(y_test, y_pred, average = 'micro'))
print("recall_score:",recall_score(y_test, y_pred, average = 'micro'))
print("f1_score:" , f1_score(y_test, y_pred, average = 'micro'))
```

```python
# create bar plot
plt.figure(figsize=(10, 6))

# Decision Tree Classifier
plt.subplot(1, 3, 1)
sns.barplot(x=['Accuracy', 'Precision', 'Recall', 'F1 Score'], 
            y=[accuracy_score(y_test, y_pred), 
               precision_score(y_test, y_pred, average='micro'), 
               recall_score(y_test, y_pred, average='micro'), 
               f1_score(y_test, y_pred, average='micro')])
plt.title('Decision Tree Classifier')

# Random Forest Classifier
plt.subplot(1, 3, 2)
sns.barplot(x=['Accuracy', 'Precision', 'Recall', 'F1 Score'], 
            y=[accuracy_score(y_test, y_pred), 
               precision_score(y_test, y_pred, average='micro'), 
               recall_score(y_test, y_pred, average='micro'), 
               f1_score(y_test, y_pred, average='micro')])
plt.title('Random Forest Classifier')

# XGBoost Classifier
plt.subplot(1, 3, 3)
sns.barplot(x=['Accuracy', 'Precision', 'Recall', 'F1 Score'], 
            y=[accuracy_score(y_test, y_pred), 
               precision_score(y_test, y_pred, average='micro'), 
               recall_score(y_test, y_pred, average='micro'), 
               f1_score(y_test, y_pred, average='micro')])
plt.title('XGBoost Classifier')

plt.tight_layout()
plt.show()
```
